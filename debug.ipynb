```python
import numpy as np
import math

#inits weights and biases + activation functions
w_1 = np.random.rand(2,2)
w_2 = np.random.rand(1,2)
b_1 = np.random.rand(2)
X_train = np.ones((2,2))
X_train[:,0] = np.array([1,2])
X_train[:,1] = np.array([2,3])
Y_train = np.array([1,0])

def relu(x):
    return (x>0) * x
    
def drelu(x):
    return x > 0

def sigmoid(x):
    return 1/(1 + (math.e)**(-x))

def dsigmoid(x):
    return sigmoid(x) * (1-sigmoid(x))

def feedforward(activation, w, b, a):
  z_out = np.matmul(w,a)
  z_out += b.reshape(b.size)
  n = z_out.size
  a_out = np.copy(z_out)
  if activation == "sigmoid":
    for i in range(n):
      a_out[i] = sigmoid(z_out[i])
  elif activation == "relu":
    for i in range(n):
      a_out[i] = relu(z_out[i])
  elif activation == "none":
    return z_out,a_out
  else:
    return "Unrecognized activation function"
  return z_out,a_out
      
def cost(a, i):
    return (a) ** 2
  
def backprop(activation,z,a,w,dz):
  dw = np.matmul(dz.reshape((dz.size,1)),a.reshape((1,a.size)))
  #print(dw)
  dz_out = np.matmul(dz.reshape((1,dz.size)),w)
  n = dz_out.size
  if activation == "sigmoid":
    for x in range(n):
      dz_out[0][x] = dz_out[0][x] * dsigmoid(z[x])
  elif activation == "relu":
    for x in range(n):
      dz_out[0][x] = dz_out[0][x] * drelu(z[x])
  else:
    return dz_out.reshape(n),dw
  return dz_out.reshape(n),dw


def relu_epoch(alpha,w_1, w_2, b_1):
  for i in range(2):
    z_1,a_1 = feedforward("relu",w_1,b_1,X_train[:,i])
    z_2,a_2 = feedforward("relu",w_2,np.zeros(1),a_1) 
    
    da_2 = a_2 - Y_train[i]
    dz_2 = da_2 * drelu(a_2)
    dz_1, dw_2 = backprop("relu",z_1,a_1,w_2,dz_2)
    dz_0, dw_1 = backprop("none",X_train[:,i],X_train[:,i],w_1,dz_1)
    
    w_1 -= alpha * dw_1
    w_2 -= alpha * dw_2
    b_1 -= alpha * dz_1 
    #print(w_1,w_2,b_1)
  return w_1,w_2,b_1
```
